```python
import os
from dotenv import load_dotenv
from firecrawl import FirecrawlApp

# 1. Load the hidden variables from the .env file
load_dotenv()

# 2. Safely grab the API key
# os.getenv will look into the loaded .env file and pull the value matching the name
my_key = os.getenv("FIRECRAWL_API_KEY")

# 3. Initialize Firecrawl with the secured key
app = FirecrawlApp(api_key=my_key)


# 4. Run your scrape!
result = app.scrape(
    'https://valmikiramayan.net/',
    formats=['markdown']
)

print(result.markdown)
```
